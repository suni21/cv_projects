```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from mtcnn import MTCNN

%matplotlib inline
```

```python
!wget https://raw.githubusercontent.com/suni21/cv_projects/main/people.jpg
```

```python
image = cv2.imread("people.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
mtcnn = MTCNN()
```

```python
faces = mtcnn.detect_faces(image)
draw_rect = np.copy(image)

for face in faces:
    x, y, w, h = face['box']
    draw_rect = cv2.rectangle(draw_rect, (x, y), (x+w, y+h), (255, 255, 0), 2)

plt.imshow(draw_rect)
```

```python
faces = mtcnn.detect_faces(image)
blurred = np.copy(image)

for face in faces:
    x, y, w, h = face['box']
    blurred[y:y+h, x:x+h] = cv2.blur(blurred[y:y+h, x:x+h], (15, 15))

plt.imshow(blurred)
```
