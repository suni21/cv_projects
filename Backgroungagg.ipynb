```python
background = cv2.imread("backgroung_change.jpg")
background = cv2.cvtColor(background, cv2.COLOR_BGR2RGB)
plt.imshow(background)
```

```python
upper_white = (168, 255,200)
lower_white = (0,0,0)
mask = cv2.inRange(image, lower_white, upper_white)

plt.imshow(mask, cmap='gray')
```

```python
background = cv2.resize(background,(image.shape[1], image.shape[0]))

background.shape, image.shape
```

```python
b = cv2.bitwise_and(background, background, mask=mask)
a = cv2.bitwise_and(image, image, mask=~mask)

plt.imshow(b)
plt.show()
plt.imshow(a)
plt.show()

endreustlt_image = cv2.add(a, b)
plt.imshow(endreustlt_image)
plt.show()
```
